## Secção 2

Para os teste será usada a variável subtype 

```python
unified_meta.columns.get_loc('SUBTYPE') #saber a coluna onde se encontra o atributo SUBTYPE 
```

Fazer a standartização dos dados 

```python
from sklearn import preprocessing

data_test=data.transpose() #Fez-se a transposta dos dados uma vez que os pacientes se encontravam nas colunas e precisavam de se encontrar nas linhas, como nos metadados

outputs = unified_meta['SUBTYPE'].values

data_scaled = preprocessing.scale(data_test)
print("Media: ", data_scaled.mean())
print("Desvio padrao: ", data_scaled.std())
```

```python
input_data = data_test.values  # apenas a matriz de valores numpy, necessário para calcular a variância
print(input_data)
```

```python
variances = input_data.var(axis = 0) # calcula a variância de cada gene
print(variances.shape)
medvar = variances.mean()
medvar
```

O objetivo da realização dos testes de estatistica univariada têm como objetivo diminuir a dimensionalidade das amostras mas mantendo a relevâncias dos dados. Para isso realizou-se testes de Variance Thresold e ANOVA

###Teste de Variance Treshold

```python
# filtros variabilidade (flat pattern)
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold= medvar*3)
input_filt = sel.fit_transform(input_data)
input_filt.shape
```

Com testes de flat pattern foi possível ficar apenas com 356 genes mais significativos para o estudo

### Teste ANOVA

```python
from sklearn.feature_selection import SelectPercentile, f_classif
import numpy as np

# fclassif - ANOVA
# seleciona genes com menor p-value (os 10% com menor valor)
selector = SelectPercentile(f_classif, percentile=10)

output_ANOVA = unified_meta.values[:,48]

selector.fit_transform(data_test, output_ANOVA)
genes_sel = np.where(selector.pvalues_ <0.01)
print(data_test.columns[genes_sel])
```

Com teste de ANOVA de 10% foi possível diminuir o número de genes para 10497

 A diferença de número de genes que resultaram dos testes estatisticos usados pode advir da metodologia que cada um usa para proceder a filtragem de genes com significado estatistico e assim signficado clinico.
 Os filtros de variabilidade procede a comparação da variância de cada coluna com os genes com uma variância média já calculada anteriormente, denominada de medvar. Assim, sendo este o threshold, todas as colunas com menor variância que este threshold, certamente, serão muito semelhantes e não serão alvo de atenção clínica para o escrutinio desta condição clínica.
 Já o teste ANOVA teve o número de genes para 10497 ao invés dos 356. Aqui, procede-se a uma análise de variância para comparar médias de populações independentes. Com base na variância de grupos de genes selecionados, procede-se a uma comparação das variâncias das médias entre os diferentes grupos de genes selecionados pela ANOVA e, assim, reduzir a dimensionalidade da amostra.
 Desta forma, este último aparenta um maior escrutinio pois não usa threshold na análise da  variância.


###PCA


```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

n=15
pca = PCA(n_components=n)
pca.fit(data_scaled)
x_reduced = pca.transform(data_scaled)


plt.bar(range(n), pca.explained_variance_ratio_*100)
plt.xticks(range(n), ['PC'+str(i) for i in range(1,n+1)])
plt.title("Variância explicada por PC")
plt.ylabel("Percentagem")
plt.show()
```

O uso do PCA prendeu-se com o objetivo de procurar um grupo de genes que explicasse a maioria da variância entre os valores das amostras. No nosso caso, os valores são muito similares e com percentagem baixa, não podendo ser retiradas nenhumas conclusões sobre o grupo de genes mais significativo do estudo

Pela observação do gráfico se encontram todos próximos mas o HIGH E LOW apresentam maior dispersão relativamente aos outros. Isto poderá indicar a existência de outliers dentros destes grupos. 

```python
# clustering hierarquico
from sklearn.cluster import AgglomerativeClustering
import pandas as pd

hclust = AgglomerativeClustering(n_clusters=4, affinity="euclidean",
                                 linkage="ward") #A criação de 4 clusters é devido a serem 4 variáveis
hclust.fit(data_scaled)

output_cluster = unified_meta.values[:,48] 
res_hc = hclust.labels_
pd.crosstab(output_cluster, res_hc)
```

É possível perceber que o cluster 1 apresenta afinidade para o subtipo HIGH o que poderá indicar que este é o subtipo mais distinto de entre os restantes.
Isto poderá indicar que os genes aqui agrupados apresentam elevada semelhança entre eles para o subtipo HIGH e o contrário para os restantes.
Dos restantes clusters, não se consegue concluir nada significante, apenas que poderão conter genes semelhantes entre todos os tipos.

```python
# k-means
from sklearn.cluster import KMeans

kmeans_est = KMeans(init='random', n_clusters=4, n_init = 100)
res_kmeans = kmeans_est.fit(data_scaled).predict(data_scaled)

pd.crosstab(output_ANOVA, res_kmeans)
```

A diferença de clusters com o anterior não é muito persetível, no entanto no cluster 0, a proximidade entre genes do LOW e MSI é elevada comparativamente com os outros dois subtipos. 

Em ambos os testes de clustering, é possível analisar que um dos clusters apresenta pouca semelhança genética entre os dados, podendo sem agrupados os dados em 3 clusters ao invés de 4.

## Secção 3 

Os algoritmos de aprendizagem de máquina, neste caso supervisionada, permitem fazer predição e extrair relações para variáveis de saída a partir de valores de variáveis de entrada. Neste caso iremos prever o tipo de variação de tumor, em particular o seu subtipo. Esta classe apresenta uma distribuição relativamente uniforme entre os tipos, logo será bom para proceder à análise. Existem pacientes em que esse valor é nulo, pelo que só serão testados em 504 pacientes.

### Treino do modelo

Irá ser feito o treino de 2/3 dos dados e usado 1/3 para teste. Para isso é neceessário  fazer a divisão dos dados, neste caso dos pacientes em 3 de forma 

```python
import numpy as np

num_tst = int(data_test.shape[0] / 3) #divide os dados em 3 
indices = np.random.permutation(data_test.shape[0])


train_in = data_scaled[indices[:-num_tst]]
train_out = outputs[indices[:-num_tst]]
test_in  = data_scaled[indices[-num_tst:]]
test_out = outputs[indices[-num_tst:]]
```

Sendo a variável sobre o tipo de cancro discreta os testes de regressão linear/logistica não são possíveis de ser realizados

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

knn = KNeighborsClassifier()
knn= knn.fit(train_in, train_out)

preds_knn = knn.predict(test_in)

print("PECC Kvizinhos:", accuracy_score(test_out, preds_knn))

#vai prever a probabilidade de ser cada um dos 3 tipos de tumores
```

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier()
tree_model = tree_model.fit(train_in, train_out)

preds_tree = tree_model.predict(test_in)

print("PECC Arvore:", accuracy_score(test_out, preds_tree))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb_model = GaussianNB()
gnb_model = gnb_model.fit(train_in, train_out)

pred_gnb = gnb_model.predict(test_in)
print("PECC NB:", accuracy_score(test_out, pred_gnb))
```

Tendo sido usados 1/3 dos dados, os resultados de accuracy dos diversos modelos supervisionados de apredizagem de máquina apresentam probabilidades bastante semelhantes, sendo por volta dos 70%. A que apresenta melhor resultado é a árvore de decisão.
O modelos de Kvizinhos é aquele que apresenta um resultado bastante inferior as restantes

# Métodos de estimação de erros


### Validação cruzada

A validação cruzada é um processo que primeiramente procede à partição do nosso dataset em diversos subsets de dimensão semelhante, mas com uma amostra aleatória de valores do nosso dataset original.
De seguida, procede-se à divisão dos nossos subsets em subsets de treino e de teste, sendo que apenas um será usado para testar o modelo.
Depois procede-se a uma iteração percorrendo o nosso conjunto de subsets usando o modelo desejado e fazendo variar o subset que é usado para teste, e os que são usados para treino, até assegurar que todos os subsets foram usados pelo menos uma vez como conjunto de teste. 


Usamos validação cruzada com 5 folds no nosso caso, ou seja, dividimos os nossos dados em 5 conjuntos diferentes, que foram então utilizados como explicado acima.
O principal uso da validação cruzada, além de aumentar a accuracy do modelo é ajudar a prevenir cenários como overfitting, ou seja, um modelo ter uma accuracy altíssima num caso singular, mas que se revela pouco útil em modelos com dados diferentes.

```python
from sklearn.model_selection import cross_val_score

scores_knn = cross_val_score(knn, data_scaled, outputs, cv = 5)
print(scores_knn)
print(scores_knn.mean())
```

```python
scores_tree = cross_val_score(tree_model, data_scaled, outputs, cv = 5)
print(scores_tree)
print(scores_tree.mean())
```

```python
scores_gnb = cross_val_score(gnb_model, data_scaled, outputs, cv = 5)
print(scores_gnb)
print(scores_gnb.mean())
```

Como é possível analisar, apesar de ser ligeiramente inferiores, estes confirmam os resultados obtidos anteriormente.

Apesar de a percentagem estar por volta dos 70%, não consideramos uma percentagem suficientemente elevada para retirar conclusões fidedignas acerca das relações entre as variáveis, dado que o nosso dataset é reduzido (504).
