## Secção 3 

os algoritmos de aprendizagem de máquina, neste caso supervisionada, permitem fazer predição e extrair relações para variáveis de saída a partir de valores de variáveis de entrada.Neste caso iremos prever o tipo de variação de tumor.

### Realização de standartização de dados

```python
from sklearn import preprocessing


input_data = data_nova.values
output_data = unified_meta.values[:,0]

input_sc = preprocessing.scale(input_data)
print("Media: ", input_sc.mean())
print("Desvio padrao: ", input_sc.std())
print("Media: ", data_nova.mean())
print("Desvio padrao: ", data_nova.std())
```

### Treino do modelo

```python
import numpy as np

indices = np.random.permutation(len(input_data))
indices
#preciso alterar os dados
train_in = input_sc[indices[:-20]]
train_out = output_data[indices[:-20]]
test_in  = input_sc[indices[-20:]]
test_out = output_data[indices[-20:]]
```

### Os modelos a utilizar serão o K vizinhos e árvore de decisão

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(train_in, train_out)
preds_knn = knn.predict(test_in)
print("Valores previstos: ", preds_knn)
print("Valores reais: " , test_out)

#vai prever a probabilidade de ser cada um dos 3 tipos de tumores
```

```python
print(preds_knn == test_out)
sum(preds_knn == test_out) / len(preds_knn)

#eficacia do teste
```

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier()
tree_model = tree_model.fit(train_in, train_out)
preds_tree = tree_model.predict(test_in)

print("Valores previstos: ", preds_tree)
print("Valores reais: " , test_out)
```

```python
print(preds_tree  == test_out)
sum(preds_tree  == test_out) / len(preds_tree)

#eficacia do teste
```
